```python
# 1-冒泡排序(Bubble Sort)

import random

'''
输入：待排列表 A，至少包含两个元素
输出：无
结果：A 就地升序排列
'''
def bubble_sort_0(A: list):
    if len(A) <= 1:
        return
     # round bubble position: [len-1, 1]
    for i in range(len(A)-1, 0, -1):
        # compare index = [1, i]
        for j in range(1, i+1):
            # swap if prev>current in disorder
            if (A[j-1] > A[j]):
                A[j-1], A[j] = A[j], A[j-1]
        print('top={}, max={}'.format(i, A[i]))

def bubble_sort_1(A: list):
    if len(A) <= 1:
        return
    top = len(A) # exclusive
    done = False
    while not done:
        done = True
        # round bubble position: [1, count-1]
        for j in range(1, top):
            # swap if prev>current in disorder
            if (A[j-1] > A[j]):
                A[j-1], A[j] = A[j], A[j-1]
                print("    swap ({}, {}) -> {}".format(j, j-1, A))
                done = False
            else:
                print("    keep (%d, %d)" % (j, j-1))
        if not done:
            print('top={}, max={}'.format(top-1, A[top-1]))
            top -= 1 # decrease right endpoint, shrink compare-swap range
        else:
            print('top={}, no swap and break'.format(top-1))

def bubble_sort_2(A: list):
    if len(A) <= 1:
        return
    top = len(A) # exclusive as range stop
    # 冒泡位置索引范围为 range(1, top)：当 top=2 时，j=1，前向比较边界(1,0)；当 top=1 时，range为空。
    while top>1:
        swapTop = 1 # no need next swap
        # round bubble position [1, count-1]
        for j in range(1, top):
            if (A[j-1] > A[j]):
                A[j-1], A[j] = A[j], A[j-1] # swap
                print("    swap ({}, {}) -> {}".format(j, j-1, A))
                # record as next top expectation
                swapTop = j
            else:
                print("    keep (%d, %d)" % (j, j-1))
        # 打印本轮冒泡位置和下一轮冒泡位置
        print('top={},{}, next swapTop={}'.format(top-1, A[swapTop], swapTop-1))
        # 本轮如果没交换swapTop=1；如果有交换，swapTop为本轮冒泡位置，下次冒泡位置缩小到swapTop-1。
        top = swapTop # adjust next top, shrink compare-swap boundary

if __name__ == "__main__":
    # 在 [1,10] 之间随机挑选6个数
    n = 6; left = 1; right = 10
    iList = random.sample(range(left,right+1), n)
    iList1 = iList.copy() # iList[:]
    print('iList1 =', iList1)
    bubble_sort_1(iList1)
    print('sorted iList1 =', iList1)
    print('-'*40)
    iList2 = iList.copy() # iList[:]
    print('iList2 =', iList2)
    bubble_sort_2(iList2)
    print('sorted iList2 =', iList2)
    print('-'*80)
    # 重复部分元素
    # random.seed(10)
    cList = random.sample(iList, k=10, counts=[1,2,3,1,2,1])
    cList1 = cList.copy()
    print('cList1 =', cList1)
    bubble_sort_1(cList1)
    print('1 sorted cList1 =', cList1)
    print('-'*40)
    # 测试不同算法
    # random.seed(10)
    # cList = random.sample(iList, k=10, counts=[1,2,3,1,2,1])
    cList2 = cList.copy()
    print('cList2 =', cList2)
    bubble_sort_2(cList2)
    print('1 sorted cList2 =', cList2)
    print('-'*80)
    # 测试用例：最左侧、最右侧部分已经排好序。
    A = [2, 3, 5, 9, 4, 6, 8, 9, 10, 10]
    A1 = A.copy()
    print('A1 =', A1)
    bubble_sort_1(A1)
    print('1 sorted A1 =', A1)
    print('-'*40)
    A2 = A.copy()
    print('A2 =', A2)
    bubble_sort_2(A2)
    print('2 sorted A2 =', A2)
    print('-'*80)
    # 测试用例：完全逆序。
    A3 = [29, 28, 27, 26, 25, 24, 23, 22, 21, 20]
    print('A3 =', A3)
    bubble_sort_2(A3)
    print('sorted A3 =', A3)
    print('-'*40)
    # 测试用例：初始有序。
    print('A3 =', A3)
    bubble_sort_2(A3)
    print('sorted A3 =', A3)
```

```python
# 2-选择排序(Selection Sort)

import random

'''
输入：待排列表 A，至少包含两个元素
输出：无
结果：A 就地升序排列
'''
def selection_sort(A: list):
    if len(A) <= 1:
        return
    # minPos: round smallest index: [0, len(A)-2]
    for i in range(0, len(A)-1):
        rightSlice = A[i:]
        minValue = min(rightSlice)
        # minIndex = A.index(minValue) # 可能有重复元素
        minIndex = rightSlice.index(minValue) + i # 注意偏移
        # swap minIndex to rank/minPos
        if i != minIndex:
            # A[minIndex] = A[i]
            # A[i] = minValue
            A[i], A[minIndex] = A[minIndex], A[i]
            print("swap({}, {}) -> {}".format(minIndex, i, A))
        else:
            print("keep %d as min" % i)

if __name__ == "__main__":
    # 在 [1,10] 之间随机挑选6个数
    n = 6; left = 1; right = 10
    iList = random.sample(range(left,right+1), n)
    print('iList =', iList)
    selection_sort(iList)
    print('sorted iList =', iList)
    print('-'*40)
    # 重复部分元素
    cList = random.sample(iList, k=10, counts=[1,2,3,1,2,1])
    print('cList =', cList)
    selection_sort(cList)
    print('sorted cList =', cList)
    print('-'*40)
    # 测试用例：最左侧、最右侧部分已经排好序。
    A = [2, 3, 5, 9, 4, 6, 8, 9, 10, 10]
    print('A =', A)
    selection_sort(A)
    print('sorted A =', A)
    print('-'*80)
    # 测试用例：完全逆序。
    A3 = [29, 28, 27, 26, 25, 24, 23, 22, 21, 20]
    print('A3 =', A3)
    selection_sort(A3)
    print('sorted A3 =', A3)
    print('-'*40)
    # 测试用例：初始有序。
    print('A3 =', A3)
    selection_sort(A3)
    print('sorted A3 =', A3)
```

```python
# 3.插入排序(Insertion Sort)

import random

'''
输入：待排列表 A，至少包含两个元素
输出：无
结果：A 就地升序排列
'''
# 对应 insertionSort0（整体后移）
def insertion_sort_0(A: list):
    if len(A) <= 1:
        return
    for i in range(1, len(A)): # i∈[1, len(A)-1]
        e = A[i] # 待插基数
        # 顺序遍历基数前的有序数列，寻找合适插入位置
        for j in range(0, i): # [0, i-1]
            if e <= A[j]: # 重复左插
                # [j+1, i] <- [j, i-1]
                A[j+1:i+1] = A[j:i] # 腾出j
                # 将基数插入合适位置
                A[j] = e
                print('i={}, j={} -> {}'.format(i, j, A))
                break

# 基于 insertion_sort_0 改为 while 循环查找插入索引，再整体后移
def insertion_sort_1(A: list):
    if len(A) <= 1:
        return
    for i in range(1, len(A)): # i∈[1, len(A)-1]
        e = A[i] # 待插基数
        j = i-1 # 基数前的有序数列起始索引
        # 逆序遍历基数前的有序数列，寻找合适插入位置
        while e<A[j] and j>=0:
            j -= 1
        # 最后一次满足e<A[j]后，j--变成左边界，应插j+1位置
        if j+1 != i: # skip if j remains i-1
            # [j+2, i] <- [j+1, i-1]
            A[j+2:i+1] = A[j+1:i] # 腾出j+1
            A[j+1] = e # 重复右插
            print('i={}, j+1={} -> {}'.format(i, j+1, A))

# 对应 insertionSort4（毗邻交换）
def insertion_sort_2(A: list):
    if len(A) <= 1:
        return
    for i in range(1, len(A)): # i∈[1, len(A)-1]
        j = i # 基数索引
        # 前向比较，逆序交换
        while A[j-1]>A[j] and j>0:
            A[j-1],A[j] = A[j],A[j-1]
            print('i={}, j={}, swap backward -> {}'.format(i, j, A))
            j -= 1

if __name__ == "__main__":
    # 在 [1,10] 之间随机挑选6个数
    n = 6; left = 1; right = 10
    iList = random.sample(range(left,right+1), n)
    print('iList =', iList)
    insertion_sort_0(iList)
    print('sorted iList =', iList)
    print('-'*40)
    # 重复部分元素
    cList = random.sample(iList, k=10, counts=[1,2,3,1,2,1])
    print('cList =', cList)
    insertion_sort_0(cList)
    print('sorted cList =', cList)
    print('-'*40)
    # 测试用例：最左侧、最右侧部分已经排好序。
    A = [2, 3, 5, 9, 4, 6, 8, 9, 10, 10]
    print('A =', A)
    insertion_sort_2(A)
    print('sorted A =', A)
    print('-'*80)
    # 测试用例：完全逆序。
    A3 = [29, 28, 27, 26, 25, 24, 23, 22, 21, 20]
    print('A3 =', A3)
    insertion_sort_2(A3)
    print('sorted A3 =', A3)
    print('-'*40)
    # 测试用例：初始有序。
    print('A3 =', A3)
    insertion_sort_2(A3)
    print('sorted A3 =', A3)
```

```python
# 4-希尔排序(Shell Sort)

import random

'''
输入：待排列表 A，至少包含两个元素
输出：无
结果：A 就地升序排列
'''
def shell_sort(A: list):
    if len(A) <= 1:
        return
    nLen = len(A)
    offset = nLen // 2 # 初始间隔
    while offset > 0:
        print("--------------- offset = %d ---------------" % offset)
        for i in range(offset, nLen): # [offset, nLen-1]
            print("--------------- right = %d" % i)
            for k in range(i, offset-1, -offset):
                if A[k] < A[k-offset]:
                    A[k],A[k-offset] = A[k-offset],A[k] # swap
                    print("    swap (%d, %d)" % (k, k-offset))
                else:
                    print("    keep (%d, %d)" % (k, k-offset))
            print('---------------', A)
        offset //= 2

if __name__ == "__main__":
    # 在 [1,10] 之间随机挑选6个数
    n = 6; left = 1; right = 10
    iList = random.sample(range(left,right+1), n)
    print('iList =', iList)
    shell_sort(iList)
    print('sorted iList =', iList)
    print('-'*40)
    # 重复部分元素
    cList = random.sample(iList, k=10, counts=[1,2,3,1,2,1])
    print('cList =', cList)
    shell_sort(cList)
    print('sorted cList =', cList)
    print('-'*40)
    # 测试用例：最左侧、最右侧部分已经排好序。
    A = [2, 3, 5, 9, 4, 6, 8, 9, 10, 10]
    print('A =', A)
    shell_sort(A)
    print('sorted A =', A)
    # 测试用例：完全逆序。
    A3 = [29, 28, 27, 26, 25, 24, 23, 22, 21, 20]
    print('A3 =', A3)
    shell_sort(A3)
    print('sorted A3 =', A3)
    print('-'*40)
    # 测试用例：初始有序。
    print('A3 =', A3)
    shell_sort(A3)
    print('sorted A3 =', A3)
```
